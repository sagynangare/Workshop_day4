# Important topics/concepts related to NLP:

**1. Corpus:** It is collection of text document, corpus--->document--->Paragraphs--->sentences 

**2. Tokens:** It is a smaller unit of the sent(words, phrases etc)

**3. N-grams:** It is the combination of word/character together

sentence='I love my phone'
**Unigram(N=1):** 'I','love','my','phone'
**Bigram(N=2):** 'I love','love my','my phone'
**Trigram(N=3):**'I love my','love my phone'

**4. Tokenization:** It is the process of splitting a text into smaller units(tokens)
 
  **Techniques in Tokenization:**
  
  **a. White space tokenizer/Unigram tokenizer: We split our text by white spaces**
  
  **b. Regular Expression Tokenizer:**
      sent="football, cricket, Golf, Tennis"
      re.split(r'[,\s]',sent)
      tokens="football" ,"cricket","Golf","Tennis"

**5. Normalization: We convert word to its base form**
    structure of token: <prefix><morpheme><suffix>
    
    ex: Antinationalist
    Anti + national +list
    
    -This technique is useful to perform data dimensionality and text cleaning
   
    
**Types Of Normalization**
      ** Stemming:**t is used to remove inflectional form of the token
    
    Form       suffix    stem
    studies     -es       studi
    studying    -ing      study
    
    laughing, laughed, laughs, laugh---> laugh
    
    
    
    
      **2. Lemmatization: It is systematic process for reducing a tokens to its lemma**
    eg:
        running, ran, run, runs----> run
        am, is,are ---> be
    
        if we take running as verb--->run
        if we take running as noun-->running

**6. Stopwords:** These are the words which are filtered out before or after processing of natural language data

**7.Part of Speech tag(POS tag):** It is the process of assinging one of the parts of speech to the given word based on its defination & its context.

**8.Bag of words(BOW):** It is a way of representing text data when modeling text with ML. BOW is a way of extracting features from the text data for use in modeling.

**9. TF-IDF (Term Frequency vs Inverse Document Frequency)

#              ```No of repetition of word in sentence
#  TF = ------------------------------------------------------
#            Total No of words in sentence```


#   ```              Total no of sentences
# IDF = log(-------------------------------------------------)
#                  No of sentence containing word```


sent1 =   good   boy
sent2 =   good   girl
sent3 =   boy    girl  good

words        Frequency
good           3
boy            2
girl           2

# TF

       sent1      sent2     sent3
 good  1/2         1/2       1/3
 boy   1/2          0        1/3
 girl   0           1/2      1/3
 
# IDF

    words        IDF
    good        log(3/3)=0
    boy         log(3/2)
    girl        log(3/2)
    

## tf_idf = TF * IDF
            f1     f2                  f3
           good    boy                 girl
    sent1   0      1/2*(log(3/2))       0
    sent2   0       0                 1/2*(log(3/2))
    sent3   0    1/3*(log(3/2))       1/3*(log(3/2))
